# 猜数游戏

用学习过的技术来完成一个简单的猜数游戏，本案例综合应用了前面所学的知识，展示了如何使用多个类，以及多个类之间的协作。

注意：我们之前区分了类和测试类，本例子中涉及多个类，请仔细体会各个类的功能，以及他们之间是如何组合在一起发挥作用的。

## 游戏功能

这个游戏涉及两个对象，三个类:

1. `game`对象是`GuessGame`类的实例，产生0-9之间的随机数，作为待猜测的数。
2. `player`对象是`Player`类的实例会猜测game对象生成的数
3. `GameLauncher`类是主类，用于建立和装配对象

## 程序逻辑

![guessgame](img/guessGame.png)
1. GameLauncher类有main（）方法, 是应用程序的入口
2. 在main（）方法中，创建一个GuessGame对象，并调用其startGame（）方法。
3. GuessGame对象的startGame（）方法是整个游戏进行的地方。 它创建玩家(player)对象，然后生成一个随机数（玩家猜测的目标），之后，要求每个玩家(player)猜测，并检查结果，若玩家猜测正确，输出获胜信息，若猜测错误，再次猜测。

代码如下：


```java
// 玩家类
public class Player {
	// 记录玩家猜测的数
	int number = 0;
	
	// 玩家猜数的方法，猜测一个数，保存到实例变量number中
	public void guess() {
		// 生成0-9之间的随机数，来模拟猜数
		// Math.random() 生成[0,1)之间的随机数，乘以10得到[0,10)之间的数
		number = (int)(Math.random()*10);
		System.out.println("玩家猜:" + number);
	}
}

// 游戏类
public class GuessGame {
	// 定义玩家对象
	Player p1;

	// 启动游戏
	public void startGame() {
		// 建立玩家对象
		Player p1 = new Player();

		// 生成一个随机数[0,10)（玩家猜测的目标）
		// Math.random() 生成[0,1)之间的随机数，乘以10得到[0,10)之间的数
		int targetNumber = (int) (Math.random() * 10);
		
		// 使用循环，猜测正确才退出循环
		while (true) {
			System.out.println("目标数是：" + targetNumber);
			
			// 玩家猜测
			p1.guess();
			
			// 检查玩家是否猜测正确
			if (p1.number == targetNumber) {
				System.out.println("猜测正确，你赢了!");
				break; // 退出循环
			} else {         
				System.out.println("猜测错误，继续尝试");
			}
		}
	}
}


public class GameLauncher {
	// 是应用程序的入口
	public static void main(String[] args) {
		// 建立game对象
		GuessGame game = new GuessGame();
		// 启动游戏
		game.startGame();
	}
}

// 运行程序
GameLauncher.main(null);
```
